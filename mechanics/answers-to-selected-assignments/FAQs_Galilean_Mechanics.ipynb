# Galilean assumptions and inertial observers

##Consequences of Galilean Space-Time Assumptions

Galilean space-time assumptions imply that:

* A universal clock exists and hence time is like $\mathbb{R}$
* Space is Euclidean and an arbitrary orthonormal frame $\mathbf{e}$ can be constructed and hence points in space can be identified with elements of $\mathbb{R}^3$
* The Euclidean inner product in $\mathbb{R}^3$ defines distances between points and angles between lines.


So basically the Galilean space-time assumptions imply that space-time is like $\mathbb{R}\times\mathbb{R}^3$ and that the Euclidean inner product in $\mathbb{R}^3$ defines distances between points and angles between lines.

## What is Euclidean Geometry?

Euclidean geometry is the study of points, lines, angles, and shapes in flat space using Euclid's axioms (including that parallel lines never meet).

It is also the geometry where that interior angles of a triangle sum upto $\pi$.

In curved spaces the interior angles of a triangle do not necessarily sum upto $\pi$.
For example on a sphere the sum of interior angles of a tringle is always more than $\pi$. The Euclidean Pythogorian theorm, $c^2=a^2+b^2$, also fails and has to be replaced by the following. For a right spherical triangle, i.e. a triangle on the surface of a sphere whose one angle is $\frac{\pi}{2}$,
let the sides be $a, b$ and the hypotenuse be $c$. Then if the sphere's radius is $r$ the equivalent (spherical) Pythagorian theorem is:
\begin{align*}
\cos{\left(\frac{c}{r}\right)}&=\cos{\left(\frac{a}{r}\right)}\cos{\left(\frac{b}{r}\right)}
\end{align*}

## What is the Euclidean inner product in $\mathbb{R}^3$

The Euclidean inner product (also called the dot product) in $\mathbb{R}^3$ is the standard way of defining an inner product on $\mathbb{R}^3$.

For two elements
$u = (u_1, u_2, u_3)\in\mathbb{R}^3, \quad {v} = (v_1, v_2, v_3)\in \mathbb{R}^3$
the Euclidean inner product is defined as:
\begin{align*}
\langle\langle {u}, {v} \rangle\rangle = u_1 v_1 + u_2 v_2 + u_3 v_3
\end{align*}
This can also be written equivalently as the matrix product:
\begin{align*}
\langle\langle {u}, {v} \rangle\rangle = {v}^T {u}
\end{align*}
where $u,v$ are the column matrix version of the ordered tripples.

It satisfies the standard properties of an inner product:

	1.	Linearity in the first argument:
\begin{align*}
\langle\langle a{u} + b{w}, {v} \rangle\rangle = a\langle\langle {u}, {v} \rangle\rangle + b\langle\langle {w}, {v} \rangle\rangle
\end{align*}

	2.	Symmetry:
\begin{align*}
\langle\langle {u}, {v} \rangle\rangle = \langle\langle {v}, {u} \rangle\rangle
\end{align*}

	3.	Positive definiteness:
\begin{align*}
\langle\langle {u}, {u} \rangle\rangle = u_1^2 + u_2^2 + u_3^2 \ge 0,
\end{align*}
and equality holds iff ${u} = {0}$.

Geometrically, the inner product measures magnitude of an element
via:
\begin{align*}
\|u\|&=\sqrt{\langle\langle {u}, {u} \rangle\rangle}
\end{align*}
and how much two elemtns “point in the same direction,” and relates to the angle $\theta$ between them via:
\begin{align*}
\langle\langle {u}, {v} \rangle\rangle = \|{u}\| \, \|{v}\| \cos \theta
\end{align*}


For additional information on vector spaces and inner products please visit: [Vector Spaces](https://github.com/mugalan/classical-mechanics-from-a-geometric-point-of-view/blob/main/supplementary/vector_spaces.ipynb)

## Who are inertial observers

Those who agree on Galilean space-time assumptions.

##How are two inertial observers related to each other?

Conider two inertial observers $\mathbf{e}$ and $\mathbf{b}$. We will denote their respective orthonormal frames also by $\mathbf{e}$ and $\mathbf{b}$. Both observe the same space time event $A$. Let $t$ be the time of the event $A$ measured by $\mathbf{e}$ and let $\tau$ be the time of the event $A$ measured by $\mathbf{b}$.
Similarly let $x$ be the $3\times 1$ Euclidean representation of the point of the event $A$ as represented by $\mathbf{e}$ and let $\xi$ be the $3\times 1$ Euclidean representation of the point of the event $A$ as represented by $\mathbf{b}$.

Then one can show that the Galilean space time assumptions imply that
\begin{align*}
\tau=t+a
\end{align*}
where $a$ is a constant,
and
\begin{align*}\xi=\alpha+\beta t+R x\end{align*} where $\alpha,\beta$ are constant $3\times 1$ matrices and $R$ is a constant $3\times 3$ matrix.

---

The assumption that space intervals are inertial observer independent implies that $R$ is an orthogonal constant transformation (that is $R^TR=RR^T=I$). Additionally if all observers use right hand oriented frames then it can be shown that $\mathrm{det}(R)=1$ and hence in fact $R$ is a special orthogonal matrix.

Considering the origin of the two frames we can show that:
\begin{align*}
v&=\dot{o}=-R^{T}\beta=\mathrm{constant}
\end{align*}

Thus we have that the Galilean space time assumptions imply that
\begin{align*}
\tau&=t+a,\\
\xi&=\alpha +R(x-vt)
\end{align*}

---

Since the laws of mechanics deal with accelerations we see that the constants off sets $a,\alpha$ really do not play a significant role. Thus to avoid carrying them around it is typical to assume that both clocks of $\mathbf{e}$ and $\mathbf{b}$ are synchronized (that is $a=0$) and that at $t=0$ the origin of both frames coincide and hence $\alpha=0_{3\times 1}$.

Thus if a certain space-time event $A$ has the representation $(t,x)$ according to $\mathbf{e}$ then it has the representation $(t,R(x-vt))$ according to $\mathbf{b}$.

---

**Note that** $R$ here is a special orthogonal $3\times 3$ matrix. That is a $3\times 3$ matrix that satisfies the properties

\begin{align*}
R^TR&=RR^T=I_{3\times 3},\\
\mathrm{det}(R)&=1
\end{align*}
The space of all such matrices are denote by $SO(3)$.

---

Since $R$ is a constant, without loss of generality, one can always pick the orthonormal frame used by $\mathbf{e}'$ to be parallel to the one used by $\mathbf{e}$ so that $R=I_{3\times 3}$. Then we see that $\xi(t)=x(t)-vt$ in parallel translating inertial frames. **It is traditional to refer to parallel frames that translate at constant velocities with respect to each other as inertial frames.**

## What are inertial observer invariant quantities?

### Distances between symultaneous points

Consider two points in space $P$ and $Q$. Let them have the Euclidean representation $x_P=[x_{1P}\:\:x_{2P}\:\:x_{3P}]^T$ and $x_Q=[x_{1Q}\:\:x_{2Q}\:\:x_{3Q}]^T$ as represented in the $\mathbf{e}$ frame.

They have the reprepresentation $\xi_P=R(x_P-vt)$ and $\xi_Q=R(x_P-vt)$ in the $\mathbf{b}$ frame.

The distance between $P$ and $Q$ as measured by $\mathbf{e}$ is
\begin{align*}
d_{\mathbf{e}}(P,Q)=||x_Q-x_P||=\sqrt{\langle\langle (x_Q-x_P)\,,\,(x_Q-x_P) \rangle\rangle}= \sqrt{(x_Q-x_P)^T(x_Q-x_P)}
\end{align*}

Similarly in the
as measured by $\mathbf{b}$ is
\begin{align*}
d_{\mathbf{b}}(P,Q)&=||\xi_Q-\xi_P||=\sqrt{\langle\langle (\xi_Q-\xi_P)\,,\,(\xi_Q-\xi_P) \rangle\rangle}= \sqrt{(\xi_Q-\xi_P)^T(\xi_Q-\xi_P)},\\
&=\sqrt{(x_Q-x_P)^TR^TR(x_Q-x_P)}= \sqrt{(x_Q-x_P)^T(x_Q-x_P)}\\
&=d_{\mathbf{e}}(P,Q)
\end{align*}

### Angles between (simultanoeus) lines

Consider 4 points in space $A,B,P,Q$. Let them have the Euclidean representation $x_A=[x_{1A}\:\:x_{2A}\:\:x_{3A}]^T$, $x_B=[x_{1B}\:\:x_{2B}\:\:x_{3B}]^T$, $x_P=[x_{1P}\:\:x_{2P}\:\:x_{3P}]^T$ and $x_Q=[x_{1Q}\:\:x_{2Q}\:\:x_{3Q}]^T$ respectively as represented in the $\mathbf{e}$ frame.

They have the reprepresentation $\xi_A=R(x_A-vt)$, $\xi_B=R(x_B-vt)$, $\xi_P=R(x_P-vt)$ and $\xi_Q=R(x_P-vt)$ in the $\mathbf{b}$ frame.

The angle between the two lines $AB$ an $PQ$ as measured in the $\mathbf{e}$ frame is
\begin{align*}
\theta_{\mathbf{e}}&=\cos^{-1}\left(\frac{\langle\langle (x_B-x_A),(x_Q-x_P)\rangle\rangle}{||(x_B-x_A)||\:||(x_Q-x_P)||}\right)
\end{align*}
The angle between the two lines $AB$ an $PQ$ as measured in the $\mathbf{b}$ frame is
\begin{align*}
\theta_{\mathbf{b}}&=\cos^{-1}\left(\frac{\langle\langle (\xi_B-\xi_A),(\xi_Q-\xi_P)\rangle\rangle}{||(\xi_B-\xi_A)||\:||(\xi_Q-\xi_P)||}\right),\\
&=\cos^{-1}\left(\frac{(x_Q-x_P)^TR^TR(x_B-x_A)}{||(x_B-x_A)||\:||(x_Q-x_P)||}\right),\\
&=\cos^{-1}\left(\frac{(x_Q-x_P)^T(x_B-x_A)}{||(x_B-x_A)||\:||(x_Q-x_P)||}\right),\\
&=\theta_{\mathbf{e}}
\end{align*}

### Magnitude of Accelerations

Position, velocity, acceleration in the $\mathbf{e}$ frame
\begin{align*}
x &\Rightarrow \quad \mathrm{Position},\\
\dot{x} &\Rightarrow \quad \mathrm{Velocity},\\
\ddot{x} &\Rightarrow \quad \mathrm{Acceleration},
\end{align*}
Position, velocity, acceleration in the $\mathbf{b}$ frame
\begin{align*}
\xi = R(x-vt)&\Rightarrow \quad \mathrm{Position},\\
\dot{\xi} = R(\dot{x}-v)&\Rightarrow \quad \mathrm{Velocity},\\
\ddot{\xi} = R\ddot{x}& \Rightarrow \quad \mathrm{Acceleration},
\end{align*}

* This we see that, in general, position and velocity of a point is not invariant in two different inertial frames.

* Since $||\ddot{\xi}||=||R\ddot{x}||=||\ddot{x}||$ the magnitude of the accelerations are invariant in inertial frame.

* Howerver the acceleration in $\mathbf{b}$ is the 'rotated' (we will see this later) version of the acceleration in $\mathbf{e}$ frame. If $\mathbf{b}$ is parallel to $\mathbf{e}$ then $R=I_{3\times 3}$ and hence $\ddot{\xi}=\ddot{x}$ and hence accelerations are invariant for parallel inertial observers.



## When do Galilean Assumptions fail

The first evidence that there was something wrong with the Galilean assumptions came from:
* Maxwell's equations of electromagnetism, which predict that the speed of light in a vacuum is constant — independent of the motion of the observer or the source.

* The Michelson-Morley experiment (1887), which confirmed that the speed of light is the same for all inertial observers, contradicting Galilean velocity addition. See answer to Q8 in [Assignment ](https://github.com/mugalan/classical-mechanics-from-a-geometric-point-of-view/blob/main/mechanics/answers-to-selected-assignments/Assignment_InertialObservers_SampleAnswers.ipynb)

Therefore, **Galilean assumptions fail when objects move at speeds comparable to the speed of light**.

Around the same time the following two observations also contradicted Galilean mechanics

* Blackbody radiation - classical physics predicted the “ultraviolet catastrophe,” which contradicted experimental results.

* Photo electric effect - the prediction of a continuous energy spectrum violates experimental results.

These phenomena showed that **Galilean mechanics also fail at atomic and subatomic scales**.


The following two observations imply that **Galilean mechanics break down at large (astronomical) scales where gravity is strong**.

* Mercury's perihelion precession — its orbit showed a small extra shift (about 43" per century) unexplained by Newtonian gravity, hinting that the Galilean framework failed even for slow-moving planets.

* 1919 solar-eclipse observation — Eddington's measurements showed starlight bending near the Sun, confirming Einstein's prediction that gravity curves spacetime and light follows that curvature.



## How to fix the speed of light problem? (Beyond the scope of the CE1010 syllabus. Intended only for those who asked about it.)

**Fundamental Assumptions of Relativistic Space-Time**: There exists special observers, referred to as inertial observers, all of whom agree that
* space-time is 4-dimensional, continuous, homogeneous, and isotropic,
* light travels in a straight line,
* the speed of light is an inertial observer independent quantity.

This says that space-time is like $\mathbb{R}^4$. All observers who agree on these assumptions are called **Lorentzian inertial observers**.

Now instead the standard inner ptoduct a special inner product is used to distances between space time events.

Consider a relativisitc inertial observer $\mathbf{e}$. Consider two points $a=(t,x_1,x_2,x_3)\in \mathbb{R}^4$ and $a'=(t',x'_1,x'_2,x'_3)\in \mathbb{R}^4$ then the (pseudo-)inner product between these two points is defined by
\begin{align*}
\langle\langle a, a'\rangle\rangle &\triangleq tt'-\frac{1}{c^2}\left(x_1x'_1+x_2x'_2+x_3x'_3\right).
\end{align*}
This is thr famous Minkowsky metric on $\mathbb{R}^4$.

If $A$ represents the space time event denoted by $a$ and $A'$ represents the space time event denoted by $a'$ Then the distance between the two speace time events is denoted by
\begin{align*}
\mathrm{d}(A,A')&=||a'-a||=\sqrt{\langle\langle (a'-a), (a'-a)\rangle\rangle}
\end{align*}

One can now show that for two observers translating parallel to one axis their measurements are related by
\begin{eqnarray}
t' &=& \gamma \left(t - \frac{v}{c^2} {x}_i\right),\\
{x}'_i &=& \gamma \left( -vt+ {x}_i\right),\\
{x}'_j &=& {x}_j,\\
{x}'_k &=& {x}_k,
\end{eqnarray}
where
\begin{align*}
\gamma&=\left(1-\frac{v^2}{c^2}\right)^{-\frac{1}{2}}.
\end{align*}
The particular type of transformation given above is usually referred to as a boost in the $i^\mathrm{th}$ spatial direction. It can be shown that the innerproduct is invariant under these transformations.


These relations are all that is needed to derive the key results of special relativity, such as time dilation, length contraction, and relativity of simultaneity.

The mass-energy relation and energy-momentum dynamics require introducing additional structure, such as the 4-momentum or the energy-momentum tensor.

The general Lorentz transformation takes the form,
\begin{align*}
\begin{bmatrix} t' \\ {x}'\end{bmatrix}
&=
\begin{bmatrix} \gamma & \:\:& &-\frac{\gamma}{c^2}{v}^T\\
-\gamma {v} &\:\: & & (\tiny{(\gamma-1)} \frac{{vv}^T}{||{v}||^2}+I)
\end{bmatrix}
\begin{bmatrix} t \\ {x}
\end{bmatrix}
\end{align*}

#Conservation of momentum

## What are summary laws that govern the Galilean description of motion

* objects are made up of impenetrable particles that interact with each other in an observer independent manner,

* when taken pairwise particles interactions lie along the straight line that joins the two particles,

* there exists a class of observers called inertial observers who agree on the Galilean space-time assumptions,

* associated with each particle there exists an inertial observer independent quantity called mass,

* the total linear momentum of all the particles in the Universe is always conserved when viewed in an inertial frame.

## What are the consequences of conservation of linear momentum for a set of $n$ particles.

Let the center of mass of the $n$ number of particles be
\begin{align*}
\bar{x}&\triangleq \frac{\sum_{i=1}^{n}m_ix_i}{\sum_{i=1}^{n}m_i},
\end{align*}
and the total mass be $M\triangleq \sum_{i=1}^{n}m_i$ then
\begin{align*}
M\ddot{\bar{x}}= f^e,
\end{align*}
where

\begin{align*}
f^e&\triangleq \sum_{i=1}^{n} f_i^e,
\end{align*}
is the total resultant interactions the set of particles experince due to their interaction with the rest of the particles of the universe. That is, the sum of all forces the particles experience due to their interaction with the rest of the universe

This result shows that the center of mass of a collection of particles moves as if all the mass of the system were concentrated at the center of mass and acted upon by the resultant of all external forces.

## Does the conservation of momentum imply that the angular momentum of a set of isolated particles about its center of mass is conserved?

The principle of conservation only says that pair wise interactions are equal in magnitude and opposite in direction. This is not enough for the conservation of the angular momentum of a set of partticles about the center of mass to hold true. For that the additional assumption that the mutual interactions must lie along the line joining the 2 particles is also necessary.

## What does the added assumption that the mutual interactions must lie along the line joining the two particles tell us?

The total *angular momentum* of all the particle in the $\mathbf{e}$ frame about the point $O'$ (fixed or otherwise) is defined to be
\begin{align*}
\pi &\triangleq \sum_{i}^n(x_i-o)\times m_i\dot{x}_i,
\end{align*}
let the total resultant *moment of the external forces* acting on all the particles about $O'$ be
\begin{align*}
{\tau^e}\triangleq \sum_{i=1}^n(x_i-o)\times  f_{i}^e.
\end{align*}

Differentiating $\pi$ and after a bit of algebraic manipulation that invokes the assumption that mutual interactions must lie along the line joining the two particles give  

\begin{align*}
\dot{\pi}&= -M\dot{o}\times \dot{\bar{x}}+\tau^e.
\end{align*}

This expression tells us that if $O'$ is the center of mass then
\begin{align*}
\dot{\pi}&=\tau^e
\end{align*}

What this shows is that conservation of linear momentum plus the assumption that mutual particle interactions must lie along the line joining the two particles says that the total rate of change of angular momentum of a set of particles about their center of mass is equal to the total resultant *moment of the external forces about the center of mass* of all the particles.

## Are there situations where momentum is conserved even though the third law does not strictly apply?

Newton's third law states:
${f}_{12} = -{f}_{21}$.
If this holds for all internal forces in an isolated system, then total momentum $P = \sum_i m_i \dot{x}_i$ is conserved automatically.

But — the third law assumes that:
* forces act instantaneously,
* forces are equal and opposite,
* interactions are direct between particles (no field mediation).

These assumptions break down in several physical situations.

### When the third law fails but momentum is still conserved

#### (a) Electromagnetic interactions

In electrodynamics, forces between moving charges do not satisfy Newton's third law:
${f}_{12} \neq -{f}_{21}$.
Because:
* The fields propagate at finite speed of light (so action and reaction aren't simultaneous), and


However, if you include the momentum of the electromagnetic field, total momentum is conserved:
\begin{align*}
\frac{d}{dt}({P}_{\text{particles}} + {P}_\text{field}) = 0.
\end{align*}
Here,
${P}_{\text{field}} = \frac{1}{4\pi c} \int \mathbf{E} \times \mathbf{B}\, d^3x$
is the momentum stored in the fields.

So the third law fails locally, but momentum conservation still holds globally, when fields are included.


(b) Non-contact forces with delay (retarded interactions)

If one mass exerts a force that depends on the past position of another (e.g. in finite-speed gravitational or electromagnetic interactions), then
${f}_{12}(t) \neq -{f}_{21}(t)$,
because they act at different times.
Still, the total momentum of the whole system (particles + mediating field) is conserved.


(c) Variable-mass systems

In systems like rockets or jet engines, momentum conservation applies even though Newton’s third law doesn’t take its simple form.
Momentum is conserved if you include the ejected mass (exhaust gases) as part of the system:
\begin{align*}
\frac{d}{dt}(m v + m_\text{exhaust} v_\text{exhaust}) = 0.
\end{align*}


The more general and fundamental statement is Noether's theorem:

Momentum conservation arises from spatial translational symmetry of the laws of physics. So, momentum conservation is a consequence of symmetry, not necessarily of pairwise equal and opposite forces.


So in summary:

**Momentum conservation is more fundamental than Newton's third law.
The third law is just a special case of momentum conservation when forces are instantaneous and internal — but in field theories and relativistic systems, the conservation law survives even when the third law fails.**

## How are two general orthonormal frames in space related to each other?

Consider the figure above where two observers have defined two right hand oriented orthonormal frames $\mathbf{e}=[\mathbf{e}_1\:\:\: \mathbf{e}_2\:\:\: \mathbf{e}_3]$ and $\mathbf{b}=[\mathbf{b}_1\:\:\: \mathbf{b}_2\:\:\: \mathbf{b}_3]$ respectively. Let $O$ be the origin of the $\mathbf{e}$ frame and let $O'$ be the origin of the $\mathbf{b}$ frame.

Let $o=[o_1\:\:\:o_2\:\:\:o_3]^T$ be the Euclidean representation of $O'$ with respect to $\mathbf{e}$.

Observe that one can represent each of the axis $\mathbf{b}_i$ using the $\mathbf{e}'$ frame as follows.
\begin{align*}
\mathbf{b}_1= r_{11}\mathbf{e'}_1+r_{21}\mathbf{e'}_2+r_{31}\mathbf{e'}_3,\\
\mathbf{b}_2= r_{12}\mathbf{e'}_1+r_{22}\mathbf{e'}_2+r_{32}\mathbf{e'}_3,\\
\mathbf{b}_3= r_{13}\mathbf{e'}_1+r_{23}\mathbf{e'}_2+r_{33}\mathbf{e'}_3.
\end{align*}
This can be expressed in the matrix form
\begin{align*}
{[\mathbf{b}_1\:\:\: \mathbf{b}_2\:\:\: \mathbf{b}_3]}={[\mathbf{e'}_1\:\:\: \mathbf{e'}_2\:\:\: \mathbf{e'}_3]}{\begin{bmatrix}r_{11}&r_{12}&r_{13}\\
r_{21}&r_{22}&r_{23}\\
r_{31}&r_{32}&r_{33}\\
\end{bmatrix}}.
\end{align*}
Define
\begin{align*}
R&=\begin{bmatrix}r_{11}&r_{12}&r_{13}\\
r_{21}&r_{22}&r_{23}\\
r_{31}&r_{32}&r_{33}\\
\end{bmatrix}
\end{align*}
Then noting that $\mathbf{e}'$ is parallel to $\mathbf{e}$ we have in compact matrix notation
\begin{align*}
\mathbf{b}=\mathbf{e}'R=\mathbf{e}R
\end{align*}

It can be shown that $R\in SO(3)$.

---

Thus the relationship between $\mathbf{b}$ and $\mathbf{e}$ is uniquely given by
\begin{align*}
(o,R)\in \mathbb{R}^3\times SO(3)
\end{align*}

This pair can also be uniquely idenitified with the $4\times 4$ matrix

\begin{align*}
E=
\begin{bmatrix}
R & o \\ 0 & 1\end{bmatrix}\in SE(3)
\end{align*}

## How are position, velocity, and acceleration measured in two orthonormal frames related to each other?

The position of the particle $P(t)$ at a particular instant of time $t$ is described by the two observers using the Euclidean representation matrices $x=[x_1\:\:\:x_2\:\:\:x_3]^T$ and $X=[X_1\:\:\:X_2\:\:\: X_3]^T$ respectively. Then we have the following relationships:

## How can we write equations of motion for the center of mass of and object in a non-inertial frame?

Assume that $\mathbf{e}$ is an inertial frame.

Let $f$ be the representation of the total resultant external forces acting on the object as reresented in the inertial frame $\mathbf{e}$. Then
\begin{align*}
F=R^Tf
\end{align*}
is the frame $\mathbf{b}$ representation of this force.

Then Newton's equations for the center of mass motion of an object of total mass $M$ is
\begin{align*}
M\ddot{x}&=f.
\end{align*}

Using the relationship between the accelerations in the two frames we then have:
\begin{align*}
MR^T\ddot{o} +M\,\left(\widehat{\Omega}^2{X} +2\widehat{\Omega}\dot{{X}}+\dot{\widehat{\Omega}}{X}+\ddot{{X}}\right)= F.
\end{align*}
Thus this expression can be considered to be the **appropriate version of the Newton's equations in the moving frame $\mathbf{b}$.**



## What are fictitious (apparent) forces?

Imagine the situation where the observer is unaware of the motion of its frame-$\mathbf{b}$ and thinks of it as an inertial frame. Like for instance when we think of an earth fixed frame to be inertial. Then the observer, having taken a mechanics class during her undergraduate program, will interpret mass times acceleration measured in her reference frame $\mathbf{b}$ to be the force felt in $\mathbf{b}$. That is, she will think that
\begin{align*}
M\ddot{{X}}&=F-\left(MR^T\ddot{o}+M\widehat{\Omega}^2 {{X}}+2M\,\widehat{\Omega}\dot{{{X}}}+M\dot{\widehat{\Omega}}{{X}}\right),
\end{align*}
is the force acting on the particle as expressed in her frame $\mathbf{b}$.
However the quantity $F=R^Tf$ is the only physically meaningful force that she feels.
Thus an observer moving with the rotating frame will, in addition to the fundamental interacting forces, feel that there exists another resultant *'apparent force'*:

\begin{align*}
F_{app} &\triangleq  
{-M\;R^T\ddot{o}}\:{-M\;\widehat{\Omega}^2 {{X}}}\:{-\:2M\;\widehat{\Omega}\dot{{{X}}}}\:{- \:M\;\dot{\widehat{\Omega}} {{X}}},
\end{align*}
simply due to its ignorance of the motion of its frame.

* The first term: $-M\;R^T\ddot{o}$ is known as the *Einstein force*,

* The second term: $-M\;\widehat{\Omega}^2 {X}$ is known as the *Centrifugal* force,

* The third term $-2M\;\widehat{\Omega}\dot{{X}}$ is known as the *Coriolis* force

* The last term:
$- M\;\dot{\widehat{\Omega}} {{X}}$ is known as the *Euler* force.

Observe that all these apparent forces have mass as a multiplicative factor.
Note that the Einstein apparent force is observed due to the *translational ignorance* of the one's reference frame while the Centrifugal, Coriolis, and Euler forces are observed due to the *rotational ignorance* of the reference frame.

In your calculations **never resort to these fictitious forces!!** Just make sure you use $M\ddot{{X}}=F-\left(MR^T\ddot{o}+M\widehat{\Omega}^2 {{X}}+2M\,\widehat{\Omega}\dot{{{X}}}+M\dot{\widehat{\Omega}}{{X}}\right)$ instead of $M\ddot{{X}}=F$

## Is gravity a real interaction?

In **Newtonian physics**, gravity is treated as a *real force* acting at a distance between two masses, given by

\begin{align*}
F = G \frac{m_1 m_2}{r^2}
\end{align*}

where $G$ is the gravitational constant.  
Objects accelerate toward each other due to this mutual attraction — a genuine interaction transmitted through a gravitational field.

---

In **Einstein’s General Relativity**, gravity is not a force but the **curvature of spacetime** caused by mass-energy.  
Objects in free fall move along geodesics — the straightest possible paths in curved spacetime.  
Thus, what we *perceive* as gravitational attraction is an **apparent (fictitious) force** arising from spacetime geometry.

Gravitational waves, first detected directly in 2015 (event GW150914 by LIGO), are **oscillations in spacetime curvature** produced by accelerating massive bodies, confirming that spacetime itself is dynamic and that Einstein's geometric description of gravity is physically real.

From the perspective of General Relativity, all objects in the universe — planets, stars, and even the Sun itself — are in continuous **free fall along geodesics in curved spacetime**, following the natural geometry created by the distribution of mass and energy.

## How to write the angular momentum of an object in a non-inertial frame?

Using the coordinate transformation between $\mathbf{e}$ and $\mathbf{b}$ we chan show that the angular momentum of an point object $P_i$ about $O'$ is
\begin{align*}
\pi_i&= m_i(x_i-o)\times \dot{x}_i=R\left(\mathbb{I}_i\Omega+m_iX_i\times( R^T\dot{o}+ \dot{X}_i)\right),
\end{align*}
where
\begin{align*}
\mathbb{I}_i &\triangleq -m_i\widehat{X}_i^2=m_i\left(||X_i||^2I_{3\times 3} - X_iX_i^T\right),
\end{align*}
is defined as the **moment of inertia** of the particle $P_i$ about the point $O'$ expresssed in the frame $\mathbf{b}$. Refer to [this note book](https://github.com/mugalan/classical-mechanics-from-a-geometric-point-of-view/blob/main/mechanics/answers-to-selected-assignments/moment_of_inertia.ipynb) for additional information.

The above expression for $\pi_i$ also shows that
\begin{align*}
\Pi_i\triangleq \left(\mathbb{I}_i\Omega+m_iX_i\times(\dot{X}_i+ R^T\dot{o})\right),
\end{align*}
is the representation of the angular momentum of $P_i$ about $O'$ in the moving frame $\mathbf{b}$.

---

If the particle $P$ is fixed with respect to the moving frame
\begin{align*}
\pi_i&= R\left(\mathbb{I}_i\Omega+m_iX_i\times R^T\dot{o}\right).
\end{align*}
Summing them over all such particles in the object we get
\begin{align*}
\pi&= R\left(\mathbb{I}\Omega+M\bar{X}\times R^T\dot{o}\right).
\end{align*}
where $M=\sum_{i=1}^Nm_i$, $\mathbb{I}=\sum_{i=1}^N\mathbb{I}_i$, and $\bar{X}=(\sum_{i=1}^Nm_iX_i)/M$.

## How to write the rate of change of angular meomentum of an object in a non-inertial frame?

Differentiating the exression for $\pi_i$ in the moving frame $\mathbf{b}$ we find
\begin{align*}
\dot{\pi}_i&=R\left(\mathbb{I}_i\dot{\Omega}-\mathbb{I}_i{\Omega}\times \Omega+m_i\,\Omega\times X_i\times\dot{X}_i-m_iR^T\dot{o}\times \Omega\times X_i +m_i\dot{X}_i\times R^T\dot{o}+m_iX_i\times (R^T\ddot{o}+\ddot{X}_i)\right).
\end{align*}

On the other hand by differentiating $
\pi_i= m_i(x_i-o)\times \dot{x}_i$ directly we have that
\begin{align*}
\dot{\pi}_i&= -m_i\dot{o}\times \dot{x}_i+\tau_i=R\left(-m_iR^T\dot{o}\times (\widehat{\Omega}X_i+\dot{X}_i) +T_i\right),
\end{align*}
where $T_i=R^T((x_i-o)\times f_i)=X_i\times F_i$ is the moment of the force acting on $P_i$ about the point $O'$ expressed in the $\mathbf{b}$ frame. Thus combining the last two expressions we have

\begin{align*}
\mathbb{I}_i\dot{\Omega}-\mathbb{I}_i{\Omega}\times \Omega+m_i\,\Omega\times X_i\times\dot{X}_i+m_iX_i\times (R^T\ddot{o}+\ddot{X}_i)
&=T_i.
\end{align*}

---

If the particle $P$ is fixed with respect to the moving frame $\mathbf{b}$ then
\begin{align*}
\mathbb{I}_i\dot{\Omega}-\mathbb{I}_i{\Omega}\times \Omega+m_iX_i\times R^T\ddot{o}
&=T_i.
\end{align*}
Summing them over all such particles in the object we get
\begin{align*}
\mathbb{I}\dot{\Omega}-\mathbb{I}{\Omega}\times \Omega+M\bar{X}\times R^T\ddot{o}
&=T,
\end{align*}
where $T=\sum_{i=1}^NT_i$, $M=\sum_{i=1}^Nm_i$, $\mathbb{I}=\sum_{i=1}^N\mathbb{I}_i$, and $\bar{X}=(\sum_{i=1}^Nm_iX_i)/M$.

This last equation is referred to as the **Euler's rigid body rotation equations**.

## How to write the kinetic energy of an object in a non-inertial frame?

he kinetic energy $KE_i$ of a particle of mass $m_i$ is defined with respect to an inertial frame $\mathbf{e}$ and is given by the relationship
\begin{equation*}
\mathrm{KE}_i \triangleq \frac{1}{2}m_i ||\dot{x}_i(t)||^2,
\end{equation*}
Expressing this using coordinate transformation the kinetic energy of the particle expressed using the frame $\mathbf{b}$ quantities is
\begin{equation}
\mathrm{KE}_i=\frac{1}{2}\left( m_i||\dot{o}||^2+2m_i\dot{o}^TR(\widehat{\Omega} X_i+\dot{X}_i)+\Omega^T\mathbb{I}_i\Omega+2m_i\dot{X}_i^T\widehat{\Omega}X_i+m_i||\dot{X}_i||^2\right).
\end{equation}

If the particle $P$ is fixed with respect to the moving frame $\mathbf{b}$ then ${X}_i=\mathrm{constant}$ and hence
\begin{align*}
\mathrm{KE}_i&=\frac{1}{2} \left( m_i||\dot{o}||^2+2m_i\dot{o}^TR({\Omega} \times X_i)+\Omega^T\mathbb{I}_i\Omega
\right).
\end{align*}


The total kinetic energy of $N$ number of particles is then
\begin{align*}
\mathrm{KE}&=\frac{1}{2} \sum_{i=1}^N\left( m_i||\dot{o}||^2+2m_i\dot{o}^TR({\Omega} \times X_i)+\Omega^T\mathbb{I}_i\Omega
\right)\\
&=\frac{1}{2} \left( M||\dot{o}||^2+2M\,\dot{o}^TR\left({\Omega} \times \bar{X}\right)+\Omega^T\mathbb{I}\Omega
\right),
\end{align*}
where $M=\sum_{i=1}^Nm_i$, $\mathbb{I}=\sum_{i=1}^N\mathbb{I}_i$, and $\bar{X}=(\sum_{i=1}^Nm_iX_i)/M$.

---

In the special case where the origin of frame $\mathbf{b}$ is located at the center of mass of the objects we have $o=\bar{x}$ and $\bar{X}=0$ and then we have
\begin{align*}
\mathrm{KE}&=\frac{1}{2} \sum_{i=1}^N\left( m_i||\dot{o}||^2+2m_i\dot{o}^TR({\Omega} \times X_i)+\Omega^T\mathbb{I}_i\Omega
\right)\\
&=\frac{1}{2} \left( M||\dot{\bar{x}}||^2+\Omega^T\mathbb{I}\Omega
\right),
\end{align*}

What this expression implies is that the kinetic energy of a set of particles with respect to some inertial frame $\mathbf{e}$ is the sum of the **translational kinetic energy of the center of mass motion** plus the **kinetic energy of the rotational relative motion of the particles with respect to the center of mass**.

# The meaning of angular velocity

## What are some of the physical significances of $SO(3)$?

* Every $R\in SO(3)$ can be considered as a unique representation of the relative orientation between two orthonormal frames $\mathbf{e}$ and $\mathbf{b}$ by the relationship
\begin{align*}
\mathbf{b}=\mathbf{e}R.
\end{align*}

* When $R\in SO(3)$ represents the relationship between two orthonormal frames $\mathbf{e}$ and $\mathbf{b}$ by the relationship $\mathbf{b}=\mathbf{e}R$ it also serves the purpose of a coordinate transformation map as follows: Let $\gamma \in \mathbb{R}^3$ be the representation of some physical quantity in the $\mathbf{e}$ frame while  $\Gamma \in \mathbb{R}^3$ be the  representation of the same physical quantity in the $\mathbf{b}$ frame. Then their relationship is given by
\begin{align*}
\gamma=R\,\Gamma.
\end{align*}

* Any $R\in SO(3)$ can also be considered as a map that acts on a point $P$ in space to give a new point $P_R$ in the following manner. Let $\mathbf{e}$ be some frame and let $x$ be the representation of $P$ in the $\mathbf{e}$ frame.
Let $P_R$ be the point in space that has the representation $Rx$ in the $\mathbf{e}$-frame. That is let $OP=\mathbf{e}\, x$ and $OP_R=\mathbf{e}\,(R x)$.
This allows one to  consider $R\in \mathrm{SO}(3)$ as a transformation that  takes $P$ to a new point $P_R$ in space by mapping
\begin{align*}
x \to Rx
\end{align*}
and identifying $P_R$ with the point in space that has the representation $Rx$ in the $\mathbf{e}$-frame. Since the Euclidean inner product on $\mathbb{R}^3$ is invariant under the action of $SO(3)$ we se that this mapping preserves angles and lengths in space and hence is called an isometric map.

## What are some properties of the space of skew-symmetric matrices, $so(3)$ ?

Recall $\widehat{\Omega}=R^T\dot{R}\in so(3)$ is a $3\times 3$ skew-symmetric matrix. The space of $3\times 3$ skew symmetric matrices is denoted by $\mathrm{so}(3)$.

It is straightforward to see that $\mathrm{so}(3)$ is a three dimensional real vector space under matrix addition and scalar multiplication. Thus it is isomorphic to $\mathbb{R}^3$. That is, there is a one-to-one and onto correspondence between elements of $\mathbb{R}^3$ and elements of $\mathrm{so}(3)$. The isomorphism $\:\:\:\widehat{}\:\:: \mathbb{R}^3\to \mathrm{so}(3)$ that is explicitly defined by,
\begin{align*}
\widehat{\Omega}=\left[ \begin{array}{ccc} 0 & -\Omega_3 & \Omega_2 \\ \Omega_3 & 0 & -\Omega_1 \\ -\Omega_2 & \Omega_1 & 0\end{array}\right],
\end{align*}
for $\Omega =(\Omega_1,\Omega_2,\Omega_3)\in \mathbb{R}^3$ gives one such identification. It is easy to verify that $\:\:\:\widehat{}\:\:: \mathbb{R}^3\to \mathrm{so}(3)$ is linear. That is $\widehat{X+Y}=\widehat{X}+\widehat{Y}$ and $\widehat{\alpha X}=\alpha \widehat{X}$ for any $X,Y\in \mathbb{R}^3$ and $\alpha\in \mathbb{R}$.
It is also easy to directly verify that this isomorphism satisfies
\begin{align*}
\widehat{\Omega}X&=\Omega \times X,\\
\langle\langle X,Y\rangle\rangle &=-\frac{1}{2}\mathrm{trace}{(\widehat{X}\widehat{Y})},
\end{align*}
for $X,Y,\Omega\in \mathbb{R}^3$.

The isometric property of $R$ as a mapping also tells us that rotating and taking the cross product is the same as taking the cross product and then rotating. This means that the following property holds:
\begin{align*}
\widehat{RX}&=R\widehat{X}R^T.
\end{align*}

Below, using symbolic computations we will also verify that the following very useful and interesting property of $3\times 3$ skew-symmetric matrices:
\begin{align*}
\widehat{X}^2&= XX^T-||X||^2I_{3\times3}.
\end{align*}
for any $X\in \mathbb{R}^3$ and $R\in \mathrm{SO}(3)$.

## How can we prove $\widehat{RX}=R\widehat{X}R^T$ ?

For any $X,Y\in \mathbb{R}^3$.. specifically any two $3\times 1$ matrices and any $R\in SO(3)$

\begin{align*}
\widehat{RX}Y=(RX)\times Y = (RX)\times (RR^TY)= R\left(X\times (R^TY)\right)=R\left(\widehat{X}(R^TY)\right)=R\widehat{X}R^TY
\end{align*}

Since this is true for all $Y$ we have
\begin{align*}
\widehat{RX}=R\widehat{X}R^T
\end{align*}

## What is the physical significance of $\widehat{\Omega}$?

Let us consider smooth rotations that are parameterised by a parameter $t$ that we may consider to be time. Let $R(t)$ be a smooth curve in the space $\mathrm{SO}(3)$ such that $R(0)=I_{3\times 3}$. Then from the above discussion on the physical significance of $R\in SO(3)$ we see that $R(t)$ represents a smooth rigid rotation of space for all $t$. Let $\mathbf{b}(t)=\mathbf{e}R(t)$ be the corresponding rotating frame. Since $R(0)=I_{3\times 3}$ we see that $\mathbf{b}(0)=\mathbf{e}$. Let $P(t)$ be a point in space that corresponds to $P(0)$ being `rotated' by $R(t)$. That is if $X$ is the representation of the point $P(0)$ in the $\mathbf{e}=\mathbf{b}(0)$ frame then $R(t)X$ is the representation of the point $P(t)$ in the
$\mathbf{e}$-frame.
 Thus since rotations by $R$ preserve angles and lengths in space, $P(t)$ will appear to be fixed as viewed in the $\mathbf{b}(t)$ frame and will be equal to $X$. That is, the representation $X$ of the point $P(t)$ in the $\mathbf{b}(t)$ frame will not depend on $t$. Let $x(t)$ be the representation of the point $P(t)$ in the $\mathbf{e}$-frame. Then $x(t)=R(t)X$.
The velocity of the point $P(t)$ in the $\mathbf{e}$-frame is thus given by $\dot{x}(t)=\dot{R}(t)X$.
Previously we have seen that $R^T(t)\dot{R}(t)=\widehat{\Omega}(t)$ is always a skew-symmetric matrix. Thus we have that the velocity of the point $P(t)$ as expressed in the $\mathbf{e}$-frame has the representation
$\dot{x}(t)={R}(t)\widehat{\Omega}(t)X$.
Therefore using the relationship between the cross product and the hat matrix multiplication we see that the velocity of the point $P(t)$ can be expressed in the $\mathbf{e}$-frame as
\begin{align*}
\dot{x}&=\dot{R}X=R\widehat{\Omega}X=R(\Omega \times X)=(R\Omega) \times (RX)=(R\Omega) \times x
=\omega\times x,
\end{align*}
where we have used the property $R(X\times Y)=RX\times RY$ and have set $\omega\triangleq R\Omega$ in the last equality. Notice that $\omega$ is the $\mathbf{e}$-frame representation of the quantity that has the representation $\Omega$ in the $\mathbf{b}$-frame. Also notice that $||\omega||=||R\Omega||=||\Omega||$.  


Since $\omega(t) \times \omega(t)=0$ we see that all points in space that lie along the direction $\omega(t)$ as viewed in the $\mathbf{e}$-frame have zero velocity when $R(t)$ acts on them by a `rotation'.

On the other hand by the definition of the cross product in $\mathbb{R}^3$ and the last equality of the expression ----  we have
\begin{align*}
\dot{x}=||\omega||\,||x||\,\sin{\theta}\,\mathbf{n} \,
\end{align*}
where $\theta$ is the angle between $\omega$ and $x$ in the $\mathbf{e}$-frame as shown in figure below and $\mathbf{n}=(\omega\times x)/||\omega \times x||$ is an orthonormal direction segment that is both mutually perpendicular to the direction given $\omega=R\Omega$ in the $\mathbf{e}$-frame and $OP$.
Thus we see that $P(t)$ is instantaneously rotating about $\omega(t)$ as viewed in the $\mathbf{e}$-frame. Since $X$ was arbitrary we see that this is true for every point in space. Which shows that under the `rotation' by $R(t)$ every point in space is instantaneously rotating about $\omega$ with an angular rate of rotation equal to $||\omega||=||\Omega||$ as viewed in the frame $\mathbf{e}$.

**The above discussion motivates one to define $\omega(t)=R(t)\Omega(t)$ to be the angular velocity of the frame $\mathbf{b}$ with respect to the frame $\mathbf{e}$ represented in the $\mathbf{e}$-frame. We will call it the spatial angular velocity of the frame $\mathbf{b}$ with respect to $\mathbf{e}$ and since $\Omega$ is its $\mathbf{b}$-frame representation, we will call $\Omega$ the body angular velocity of the frame $\mathbf{b}$ with respect to $\mathbf{e}$.**

# How to Model Rigid Body Motion? - (Beyond the scope of CE1010)

## What is a Rigid Body and how can we charachterize it?

A collection of non co-linear interacting particles $P_1,P_2,\cdots,P_n$ where the interactions ensure that the relative distance between any two particles remain the same at all time. Such a set of particles is called a **rigid body**.

RigidBody.png

Let $\mathbf{e}$ be an inertial frame with origin $O$ and and let $\mathbf{b}(t)$ be an ortho-normal frame with origin $O'$ in which all the particles $P_i$ appear to be fixed as illustrated for example in the figure above. We will call $\mathbf{b}(t)$ the body frame. Let $OO'=\mathbf{e}o(t)$. The position of the $i^\mathrm{th}$ point $P_i$ at a time $t$, is given by $x_i(t)$ with respect to the frame $\mathbf{e}$ and by $X_i$
with respect to the body frame $\mathbf{b}(t)$. Observe that since all points on the body appear to be fixed with respect to the body frame $\mathbf{b}(t)$, the representation $X_i$ is independent of time. Therefore specifying $\mathbf{b}(t)$ amounts to specifying the configuration
of the rigid body. Since $\mathbf{b}(t)$ is uniquely related to the inertial frame $\mathbf{e}$ by the rotational matrix $R(t)\in SO(3)$, where $\mathbf{b}(t)=\mathbf{e}\,R(t)$, and the position of its origin $o(t)\in \mathbb{R}^3$, the specification of
$$(o(t),R(t))\in \mathbb{R}^3\times SO(3)$$
amounts to the unique specification of the configuration of the rigid body with respect to the inertial frame $\mathbf{e}$. Similarly any $(o,R)$ where $o\in \mathbb{R}^3$ and $R\in \mathrm{SO}(3)$ defines a unique configuration of the rigid body. Thus the configuration space of rigid body motion is $ \mathbb{R}^3\times \mathrm{SO}(3)$ where we have denoted the space of $3\times 3$ special orthogonal matrices by $\mathrm{SO}(3)$.

The pair $(o,R)\in \mathbb{R}^3 \times \mathrm{SO}(3)$ can also be identified with a unique $4\times 4$ matrix
\begin{align*}
E&=\begin{bmatrix}
R & o \\ 0 & 1\end{bmatrix}.
\end{align*}
The spcace of all such $4\times 4$ matrices are referred to as the space of special euclidean group of matrices that is denote by $SE(3)$.

Recall that the quantity $\Omega$ where $\widehat{\Omega}=R^T\dot{R}$ corresponds to an instantaneous rotation of the body about the axis $\Omega$ as expressed in the body frame $\mathbf{b}$ by an
amount equal to the magnitude $||\Omega||$ and thus that $\Omega$ can be defined to be the **body angular velocity** of the rigid body. The $\mathbf{e}$ frame version of this quantity $\omega \triangleq R\Omega$ is defined to be the **spatial angular velocity** of the rigid body.

The equation
\begin{align*}
\dot{R}&=R\widehat{\Omega}
\end{align*}
that defines angular velocity is usually referred to as the **rigidbody kinematic** equations.

Since $\widehat{\Omega}=\widehat{R^T\omega}=R^T\widehat{\omega}R$ this can also be equivalently written down as
\begin{align*}
\dot{R}&=\widehat{\omega}R.
\end{align*}

## Rigid Body Equations in an Inertial Frame $\mathbf{e}$

We have seen that, the principle of conservation of Momentum implies that, for any set of particles, rigid with respect to each other or not, the rate of change of linear momentum satisfies
\begin{align*}
\dot{p}= f^e,
\end{align*}
where
\begin{align*}
f^e&\triangleq \sum_{i=1}^{n} f_i^e,
\end{align*}
is the total resultant interactions the set of particles experince due to their interaction with the rest of the particles of the universe.

---

Furthermore we have seen that the added assumption that mutual particle interactions are central,
the rate of change of the total *angular momentum* of all the particle in the $\mathbf{e}$ frame about the point $O'$ (fixed or otherwise) is

\begin{align*}
\dot{\pi}&= -\dot{o}\times p+\tau^e,
\end{align*}
where
\begin{align*}
{\tau^e}\triangleq \sum_{i=1}^n(x_i-o)\times  f_{i}^e,
\end{align*}
is the total resultant *moment of the external forces* acting on all the particles about $O'$.

---

Recall that the total linear momentum of a set of interacting particles can be written down as
\begin{align*}
p&=M(\dot{o}+R\widehat{\Omega}\bar{X})=M(\dot{o}+\widehat{\omega}R\bar{X})=M\dot{\bar{x}},
\end{align*}
where $M=\sum_{i}m_i$, $\bar{X}=\sum_{i}m_i X_i/\sum_{i}m_i$ is the center of mass of the rigid body represented with repect to frame $\mathbf{b}$ fixed to the body (body frame).

Thus we also see that that the translational kinematics are given by
\begin{align*}
\dot{o}&=\frac{1}{M}p-\omega \times R\bar{X}.
\end{align*}

---

**Summary:**

Since $(o(t),R(t))\in \mathbb{R}^3\times SO(3)$ uniquely characterizes a rigid body the rigid body equations in an inertial frame are:

\begin{align*}
\dot{o}&=\frac{1}{M}p-\omega \times \bar{x}',\\
\dot{R}&=\widehat{\omega}R,\\
\dot{p}&=f^e,\\
\dot{\pi}&=\omega \times \bar{x}'\times p+\tau_e
\end{align*}
where
\begin{align*}
\bar{x}'(t)&=R(t)\bar{X},\\
\omega &=(\mathbb{I}_c^R)^{-1}\left(\pi-\bar{x}'\times p\right).
\end{align*}
Here
\begin{align*}
\mathbb{I}_c^R\triangleq R\mathbb{I}_cR^T=R(\mathbb{I}+M\widehat{\bar{X}}^2)R^T,
\end{align*}
where
$$
\mathbb{I}=\sum_{i=1}^n-m_i\widehat{X}_i^2
$$
is the moement of inertia of the set of particles with respect to the frame $\mathbf{b}$ and $\mathbb{I}_c^R$
is defined to be the *locked inertia tensor* in the body. It can be shown that $\mathbb{I}_c^R$ is the moment of inertia tensor of the body with respect to a frame that is parallel to $\mathbf{e}$ and  orign coinciding with the center of mass $O_c$.

## Rigid Body Equations in an Inertial Frame $\mathbf{b}$

Writing the above equations using the $\mathbf{b}$ frame we have:
\begin{align*}
\dot{R}&=R\widehat{\Omega},\\
M\ddot{o}&= f^e,\\
\mathbb{I}_c\dot{\Omega}&=\mathbb{I}_c\Omega \times {\Omega}+T^e_c.
\end{align*}
where
\begin{align*}
\mathbb{I}_c&\triangleq \mathbb{I}+M\widehat{\bar{X}}^2
\end{align*}
and
$$T^e_c\triangleq -\bar{X}\times F^e+T^e$$
with $F^e=R^Tf^e$ and $T^e=R^T\tau^e$.
