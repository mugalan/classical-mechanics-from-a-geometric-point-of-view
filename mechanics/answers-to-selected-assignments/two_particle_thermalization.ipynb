# Thermalization of a 2-particle system

In what follows we consider what happens to the total kinetic energy of a set of colliding particles that are isolated from the rest of the universe. We will assume that the particles interact with each other only when they collide with each other.
Let us first consider what happens when only two particles collide with each other. Let the two particles have a mass of $m_1$ and $m_2$ respectively. Let $\mathbf{e}$ be an inertial frame and $\mathbf{b}$ be a frame that moves parallel to $\mathbf{e}$ with origin coinciding with the center of mass of the two particles. We will call the frame $\mathbf{b}$ the center of mass frame. Let $x_1, x_2$ be the Euclidean representation of the two particles in $\mathbf{e}$ and let $X_1,X_2$ be the representation of the two particles in $\mathbf{b}$. Let $o$ be the representation of the center of mass of the particles in the $\mathbf{e}$ frame. Then $x_i=o+X_i$.
Since the origin of $\mathbf{b}$ is at the center of mass of the two particles we have
\begin{align}
m_1X_1+m_2X_2 &=0.
\end{align}
Differentiating this we have
\begin{align}
m_1\dot{X}_1+m_2\dot{X}_2&=0.
\end{align}
Thus the
total linear momentum of the particles in the $\mathbf{e}$ frame can be expressed as
\begin{align}
m_1\dot{x}_1+m_2\dot{x}_2&= (m_1+m_2)\dot{o}
\end{align}
and the total kinetic energy of the particles can be expressed as
\begin{align}
\mathrm{KE}&=\frac{m_1}{2}||\dot{x}_1||^2+\frac{m_2}{2}||\dot{x}_2||^2,\\
&=\frac{(m_1+m_2)}{2}||\dot{o}||^2+\frac{m_1}{2}||\dot{X}_1||^2+\frac{m_2}{2}||\dot{X}_2||^2.
\end{align}

Let us consider what happens in a collision. Denote by a superscript $'$ the variables that describe the motion of the particle after collision. Principle of conservation of momentum implies that
\begin{align*}
(m_1+m_2)\dot{o}&=(m_1+m_2)\dot{o}'.
\end{align*}
This shows that the **center of mass velocity is unaltered by the collision**.

---

Let us assume that the **kinetic energy is not lost in the collision**. Such collisions are called **elastic collisions**. Thus $\mathrm{KE}=\mathrm{KE}'$. This gives us
\begin{align*}
\frac{(m_1+m_2)}{2}||\dot{o}||^2+\frac{m_1}{2}||\dot{X}_1||^2+\frac{m_2}{2}||\dot{X}_2||^2 &=
\frac{(m_1+m_2)}{2}||\dot{o}'||^2+\frac{m_1}{2}||\dot{X}_1'||^2+\frac{m_2}{2}||\dot{X}_2'||^2 .
\end{align*}
Since $\dot{o}=\dot{o}'$ we have
\begin{align*}
\frac{m_1}{2}||\dot{X}_1||^2+\frac{m_2}{2}||\dot{X}_2||^2 &=
\frac{m_1}{2}||\dot{X}_1'||^2+\frac{m_2}{2}||\dot{X}_2'||^2 .
\end{align*}
From $m_1\dot{X}_1+m_2\dot{X}_2=0$ we have
\begin{align*}
\dot{X}_2&=-\frac{m_1}{m_2}\dot{X}_1,\:\:\:\:\:\:\dot{X}_2'=-\frac{m_1}{m_2}\dot{X}_1',
\end{align*}
and hence
\begin{align*}
||\dot{X}_1||^2 &=||\dot{X}_1'||^2,\:\:\:\:\:\: ||\dot{X}_2||^2 =||\dot{X}_2'||^2.
\end{align*}
This shows that, **in a perfectly elastic collision, the magnitude of the velocities of each of the particles do not change when viewed in the center of mass frame**.

---

Expression $m_1\dot{X}_1+m_2\dot{X}_2=0$ also tells us that **in the center of mass frame the two particles appear to move in a straight line through the origin**. Thus **what changes in a perfectly ellasitc  collision is only the angle of this straight line**. The cosine of the angle between this line and the velocity of the center of mass can be found from $\dot{o}\cdot(\dot{X}_2-\dot{X}_1)$
\begin{align*}
\dot{o}\cdot(\dot{X}_2-\dot{X}_1)&=\dot{o}\cdot(\dot{x}_2-\dot{x}_1)=\left(\frac{m_1\dot{x}_1+m_2\dot{x}_2}{(m_1+m_2)}\right)\cdot(\dot{x}_2-\dot{x}_1)\\
&=\frac{m_2||\dot{x}_2||^2-m_1||\dot{x}_1||^2+(m_1-m_2)\dot{x}_1\cdot\dot{x}_2}{m_1+m_2}.
\end{align*}

Let us consider the case where the two particles have been contained in space so that the particles would have been colliding elastically with the walls and with each other for a long time. Since the only thing that changes in elastic particle collisions is their relative direction of motion viewed in a center of mass frame of the two particles we can
hypothesize that the average of the angle between the center of mass motion and the motion relative to the center of mass, over time, should be zero. That is
\begin{align}
\langle \dot{o}\cdot (\dot{x}_2-\dot{x}_1)\rangle =0.
\end{align}

Here we have used the customary notation of angle brackets to mean the time average over a long period of time.
This in essence assumes that **on average there is no preferred directions of motion**. Thus it is also true that
\begin{align}
\langle \dot{x}_1\cdot \dot{x}_2\rangle=0.
\end{align}
This expresses the statistical independence and isotropy of particle velocities in equilibrium — an assumption that follows from the ergodicity or randomness of the system.

Under this assumption it follows that over a long period of time there is no correlation between the center of mass motion and the motion with respect to the center of mass. Then the above expression tells us that
\begin{align}
\left\langle \frac{m_1||\dot{x}_1||^2}{2}\right\rangle
=\left\langle \frac{m_2||\dot{x}_2||^2}{2}\right\rangle
\end{align}
Which tells us that **after many collisions we could expect that the average kinetic energy of the two particles to be the same**. When the system consists of a large number of particles, taking particles pairwise into consideration we may conclude that after many collisions the average kinetic energy of all the interacting particles become the same. This is the condition that is known as **thermal equilibrium**.

---

Summary of what we have observed above:
* The law of conservation of linear momentum implies that the center of mass motion of a set of colliding particles do not change over time.
* In addition if the particle collisions are elastic then the average kinetic energy of all the colliding particles remain constant over time. This condition is known as **thermal equilibrium**.


# Question: Since the lighter, faster particles have a higher collision frequency  as they encounter other particles more often in a unit time, does this higher frequency accelerate the rate at which they reach equilibrium? Or is the time taken to reach equilibrium determined by the efficiency of energy transfer in individual impacts, regardless of how often they occur?

You are absolutely right that in thermal equilibrium, particles of different masses have the same average kinetic energy,
$\left\langle \tfrac{1}{2} m_1 v_1^2 \right\rangle = \left\langle \tfrac{1}{2} m_2 v_2^2 \right\rangle$,
so the lighter ones indeed move faster on average.

Now, your question asks whether these faster, more collision-prone particles thermalize more quickly.

⸻

1. Collision frequency vs. relaxation rate

The collision frequency (number of collisions per unit time) certainly matters — it sets how often particles exchange energy and momentum.
Faster (lighter) particles do collide more often simply because they sweep through more volume per second.

However, the rate of approach to equilibrium (called the relaxation rate) depends not only on how often collisions happen but also on how effective each collision is in redistributing energy between particles.

⸻

2. Efficiency of energy transfer

For two particles with very different masses, each collision transfers only a small fraction of kinetic energy from one to the other.

For example, if a light particle (mass $m_1$) hits a heavy one (mass $m_2 \gg m_1$), the heavy particle’s velocity changes very little — the light particle mostly bounces back.
So even though the light particle collides often, each collision is inefficient at exchanging energy.

Conversely, collisions between particles of similar mass are much more efficient at equalizing kinetic energies — a single collision can significantly alter both speeds.

⸻

3. What really determines the thermalization time

Therefore, the thermalization timescale (the time required for average kinetic energies to equalize) is governed by a combination of:
• the collision rate (which sets how many opportunities exist for exchange), and
• the energy-transfer efficiency per collision, which depends on the mass ratio and the angular distribution of scattering.

Quantitatively, kinetic theory describes this by a relaxation time $\tau \sim 1 / (\nu \,\xi)$,
where $\nu$ is the collision frequency and $\xi$ is an efficiency factor depending on how much energy is typically exchanged per collision.

⸻

4. Physical intuition

So yes — lighter, faster particles do collide more often, which helps thermalization, but that advantage is partially offset if those collisions are less effective in transferring energy to much heavier particles.
The equilibrium is reached only when both factors — frequency and efficiency — balance out across the whole system.

⸻

5. A concise summary answer

Although lighter particles collide more frequently, the time to reach equilibrium depends both on how often collisions occur and how much energy each collision transfers.
Faster collisions increase opportunities for energy exchange, but if mass differences make those collisions inefficient, the overall relaxation rate may not be much faster.
Thus, thermalization is controlled by the product of collision frequency and energy-transfer efficiency, not by either factor alone.

⸻

# SImulation example

This simulation models the thermalization process in a simple gas composed of two types of particles — light and heavy — that collide elastically and randomly in three dimensions.

At the start, particles are assigned random velocities, and at each time step:

1.	A random pair of particles is selected to collide.
2.	The collision is treated as perfectly elastic, conserving both momentum and kinetic energy.
3.	The post-collision relative velocity is rotated to a random new direction, representing the isotropy of collisions in three dimensions.

Over many such random collisions, the system evolves toward thermal equilibrium:

*	The average kinetic energy of the light and heavy particles gradually becomes equal,
\begin{align}
\left\langle \tfrac{1}{2} m_{\text{light}} v_{\text{light}}^2 \right\rangle
=
\left\langle \tfrac{1}{2} m_{\text{heavy}} v_{\text{heavy}}^2 \right\rangle,
\end{align}
even though individual velocities fluctuate continuously.

*	This demonstrates the process of energy redistribution through collisions, leading to equipartition of kinetic energy.
*	The randomization of collision directions ensures that
\begin{align}
\langle \dot{x}_i \cdot \dot{x}_j \rangle = 0,
\end{align}
meaning particle motions become statistically uncorrelated and isotropic — the defining feature of an ergodic system in thermal equilibrium.

---

In short:

The simulation illustrates how repeated random elastic collisions between particles of different masses cause their average kinetic energies to equalize over time, realizing the condition of thermal equilibrium through microscopic energy exchange.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
N = 100
masses = np.random.choice([1.0, 5.0], size=N)
velocities = np.random.randn(N, 3) * 2.0
n_steps = 15000
sample_interval = 1000  # record data every this many collisions

def random_unit_vector():
    phi = 2 * np.pi * np.random.rand()
    costheta = 2 * np.random.rand() - 1
    sintheta = np.sqrt(1 - costheta**2)
    return np.array([sintheta * np.cos(phi), sintheta * np.sin(phi), costheta])

def collide(i, j):
    """Elastic collision between particles i and j with random post-collision direction."""
    global velocities
    m1, m2 = masses[i], masses[j]
    v1, v2 = velocities[i], velocities[j]
    v_com = (m1*v1 + m2*v2) / (m1 + m2)
    v1_rel = v1 - v_com
    direction = random_unit_vector()
    v1_rel_new = np.linalg.norm(v1_rel) * direction
    v2_rel_new = - (m1 / m2) * v1_rel_new
    velocities[i] = v_com + v1_rel_new
    velocities[j] = v_com + v2_rel_new

# Helper: running average
def running_average(data, window):
    return np.convolve(data, np.ones(window)/window, mode='valid')

# --- Simulation data collectors ---
ke_light, ke_heavy, vdot_mean = [], [], []
ke_light_samples, ke_heavy_samples, vdot_samples = [], [], []

for step in range(n_steps):
    # pick random pair to collide
    i, j = np.random.choice(N, 2, replace=False)
    collide(i, j)

    # compute instantaneous quantities
    ke = 0.5 * masses[:, None] * (velocities**2)
    ke_total = ke.sum(axis=1)
    ke_light.append(np.mean(ke_total[masses == 1.0]))
    ke_heavy.append(np.mean(ke_total[masses == 5.0]))

    # average velocity correlation ⟨v_i·v_j⟩
    vdot = np.dot(velocities, velocities.T)
    np.fill_diagonal(vdot, 0)
    vdot_mean.append(np.mean(vdot)/3.0)  # normalize by dimension

    # sample distributions occasionally
    if step % sample_interval == 0 and step > 0:
        ke_light_samples.append(ke_total[masses == 1.0])
        ke_heavy_samples.append(ke_total[masses == 5.0])
        idx = np.random.choice(N, (50, 2), replace=True)
        dots = [np.dot(velocities[a], velocities[b]) for a, b in idx if a != b]
        vdot_samples.extend(dots)

# --- Smooth signals ---
window = 200
ke_light_smooth = running_average(ke_light, window)
ke_heavy_smooth = running_average(ke_heavy, window)
vdot_smooth = running_average(vdot_mean, window)

# --- Flatten samples ---
ke_light_all = np.concatenate(ke_light_samples)
ke_heavy_all = np.concatenate(ke_heavy_samples)

# ==============================
# Plot set (1): Time evolution
# ==============================
plt.figure(figsize=(12,5))

# (a) Kinetic energy evolution
plt.subplot(1,2,1)
plt.plot(ke_light, alpha=0.3, color='C0')
plt.plot(ke_heavy, alpha=0.3, color='C1')
plt.plot(ke_light_smooth, color='C0', label='Light (avg)')
plt.plot(ke_heavy_smooth, color='C1', label='Heavy (avg)')
plt.xlabel('Collision step')
plt.ylabel('Average kinetic energy')
plt.title('Thermalization: Running Averages of ⟨K⟩')
plt.legend()
plt.grid(True)

# (b) Velocity correlation decay
plt.subplot(1,2,2)
plt.plot(vdot_mean, alpha=0.3, color='plum')
plt.plot(vdot_smooth, color='purple', label='Running average')
plt.axhline(0, color='k', linestyle='--', linewidth=0.8)
plt.xlabel('Collision step')
plt.ylabel(r'$\langle \dot{x}_i \cdot \dot{x}_j \rangle$')
plt.title('Decay of Velocity Correlation (Isotropization)')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# ==============================
# Plot set (2): Statistical distributions
# ==============================
fig, axes = plt.subplots(1, 2, figsize=(12,5))

# (a) Distribution of kinetic energies
axes[0].hist(ke_light_all, bins=40, alpha=0.6, label='Light particles', color='C0', density=True)
axes[0].hist(ke_heavy_all, bins=40, alpha=0.6, label='Heavy particles', color='C1', density=True)
axes[0].set_xlabel('Kinetic Energy')
axes[0].set_ylabel('Probability Density')
axes[0].set_title('Distribution of Kinetic Energies')
axes[0].legend()
axes[0].grid(True)

# (b) Distribution of velocity correlations
axes[1].hist(vdot_samples, bins=60, alpha=0.7, color='purple', density=True)
axes[1].axvline(0, color='k', linestyle='--', linewidth=0.8)
axes[1].set_xlabel(r'$\dot{x}_i \cdot \dot{x}_j$')
axes[1].set_ylabel('Probability Density')
axes[1].set_title('Distribution of Velocity Correlations')
axes[1].grid(True)

plt.tight_layout()
plt.show()
```
