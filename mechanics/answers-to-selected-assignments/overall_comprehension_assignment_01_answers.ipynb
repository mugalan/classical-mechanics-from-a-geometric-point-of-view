#Q1

Consider the motion of three mutually interacting particles $P_1,P_2,P_3$ of mass $m_1,m_2,m_3$ repectively. The Euclidean representation of the $i^{\mathrm{th}}$ particle in some inertial frame $\mathbf{e}$ is denoted by $x_i$ where $ i \in  \{1, 2, 3\} $. Let $M=\left(\sum_{i=1}^3m_i\right)$ be the total mass of the three particles.
The mutual particle interactions are gravitational in nature and the interaction of the $ j^{\mathrm{th}}$ particle on the $ i^{\mathrm{th}}$  particle, and expressed in the inertial frame $\mathbf{e}$, is given by
\begin{align*}
G m_j \frac{{x}_j - {x}_i}{||{x}_j - {x}_i||^3}
\end{align*}
where $ G $ is the gravitational constant.

Let $\mathbf{b}$ be a right hand oriented orthonormal frame with origin coinciding with the center of mass of the 3 particles. The relationship between $\mathbf{e}$ and $\mathbf{b}$ is given by $(o,R)\in \mathbb{R}^3\times SO(3)$ where $R$ is constant. Let $X_i$ be the representation of the $i^{\mathrm{th}}$ particle in the frame $\mathbf{b}$ for $ i \in  \{1, 2, 3\} $.


## Q1.1

If the particles are isloated from the rest of the universe show that the $\mathbf{b}$ is also an inertial frame.

###Answer

The center of mass of the three particles, $\bar{x}$, with respect to $\mathbf{e}$ is given by
\begin{align*}
\bar{x}=\frac{\sum_{i=1}^3m_ix_i}{\sum_{i=1}^3m_i}.
\end{align*}

From the principle of conservation of momentum we have
\begin{align*}
\sum_{i=1}^3m_i\dot{x}_i=M\,\dot{\bar{x}}=\mathrm{constant}.
\end{align*}

Thus the center of mass of the particles is either moving at contant velocity or is at rest with respect to $\mathbf{e}$.

Thus since $o=\bar{x}$, the frame $\mathbf{b}$ is either moving at contant velocity or is at rest with respect to $\mathbf{e}$ without rotation and hence is also an inertial frame.

##Q1.2

If the particles are isloated from the rest of the universe find the rate of change of the total angular momentum of the set of three particles in the $\mathbf{e}$ frame about the origin of the $\mathbf{b}$ frame.

##Answer

The rate of change of total angular momentum of the particles in the $\mathbf{e}$ fram with respect to the origin of $\mathbf{b}$ is
\begin{align*}
\dot{\pi} = M\dot{o}\times\dot{\bar{x}}+\tau_e,
\end{align*}
where $M=\sum_{i=1}^3m_i$.

Since the particles are isolated $\tau_e=0_{3\times 1}$ and $o=\bar{x}$ we have that
\begin{align*}
\dot{\pi} = 0_{3\times 1}
\end{align*}

##Q4.1

Consider the space of $3\times3$ skew symmetric matrices that is denoted by $so(3)$ and answer the followings:

###Q4.1.1

Show that $\dot{R}=R\widehat{\Omega}$ where $\widehat{\Omega}\in so(3)$

####Answer

Since $R^TR=I_{3\times 3}$
\begin{align*}
\dot{R}^TR+R^T\dot{R}=0_{3\times3}
\end{align*}
and hence
\begin{align*}
R^T\dot{R}=-(R^T\dot{R})^T
\end{align*}
Hence
\begin{align*}
R^T\dot{R}=\widehat{\Omega}
\end{align*}
is a skew symmetric $3\times 3$ matrix and hence $\widehat{\Omega}\in so(3)$

###Q4.1.2

Show that $so(3)$ is a vector space over $\mathbb{R}$

####Answer

Let any $\widehat{X},\widehat{Y}\in so(3)$
then since
\begin{align*}
(\widehat{X}+\widehat{Y})^T&=\widehat{X}^T+\widehat{Y}^T=-(\widehat{X}+\widehat{Y})^T
\end{align*}
and for any $\lambda \in \mathbb{R}$
\begin{align*}
(\lambda\widehat{X})^T&=\lambda\widehat{X}^T=-\lambda\widehat{X}
\end{align*}
we see that $so(3)$ is a vecotr space over $\mathbb{R}$

###Q4.1.3

Show that the set $\{\widehat{e}_1,\widehat{e}_2,\widehat{e}_3\}$ forms basis for $so(3)$ where
\begin{align*}
\widehat{e}_1=\left[\begin{array}{ccc}
0 & 0 & 0\\
0 & 0 & -1\\
0 & 1 & 0
\end{array}\right],\:\:\:\:
\widehat{e}_2=\left[\begin{array}{ccc}
0 & 0 & 1\\
0 & 0 & 0\\
- 1& 0 & 0
\end{array}\right],\:\:\:\:
\widehat{e}_3=\left[\begin{array}{ccc}
0 & -1 & 0\\
1 & 0 &0\\
0 & 0 & 0
\end{array}\right]
\end{align*}

####Answer

We see that any
\begin{align*}
\widehat{\Omega}=\left[\begin{array}{ccc}
0 & -\Omega_3 & \Omega_2\\
\Omega_3 & 0 & -\Omega_1\\
-\Omega_2 & \Omega_1 & 0
\end{array}\right]\in so(3)
\end{align*}
can be uniquely written down as
\begin{align*}
\widehat{\Omega}=\Omega_1\widehat{e}_1+\Omega_2\widehat{e}_2+\Omega_3\widehat{e}_3
\end{align*}

Hence $\{\widehat{e}_1,\widehat{e}_2,\widehat{e}_3\}$ forms basis for $so(3)$

###Q4.1.4

Let $e_1=[1\:0\:0]^T,e_1=[0\:1\:0]^T,e_3=[0\:0\:1]^T$

Show that the map $\widehat{} : \mathbb{R}^3 \mapsto so(3)$ that is defined by

$e_1\mapsto \widehat{e}_1$, $e_2\mapsto \widehat{e}_2$, $e_3\mapsto \widehat{e}_3$ is a linear isomorphism between $\mathbb{R}^3$ and $so(3)$.

####Answer

For any $X,Y\in \mathbb{R}^3$ we easily see that
\begin{align*}
\widehat{X+Y}&=\widehat{X}+\widehat{Y}\\
\widehat{\lambda X}&=\lambda\widehat{X}
\end{align*}
Hence the mapping is linear.

Given any $X\in \mathbb{R}^3$ we can write it uniquely as
\begin{align*}
X=\sum_{i=1}^3X_ie_i
\end{align*}
for $X_i\in \mathbb{R}$

Now from the linearity of the hat map we have
\begin{align*}
\widehat{X}=\widehat{\sum_{i=1}^3X_ie_i}=\sum_{i=1}^3X_i\widehat{e_i}
\end{align*}
Since $\{\widehat{e}_1,\widehat{e}_2,\widehat{e}_3\}$ is also a basis for $so(3)$ we see that $\widehat{X}$ is also unique. Thus the map is 1 to 1.

Reversing the argument we also see that it is onto.
